#### 3. Make a list of each Boolean operator&#39;s truth tables (i.e. every possible combination of Boolean values for the operator and what it evaluate ).

The `and` Boolean Operator:

The and operator takes two arguments. It evaluates to False unless both inputs are True. Truth table for and operator is:

|A|	B|	A and B|
|---|---|---|
|True|	True|	True|
|False|	True|	False|
|True|	False|	False|
|False|	False|	False|

The `or` Boolean Operator:

The value of the or operator is True unless both of its inputs are False. Truth table for or operator is:

|A|	B|	A or B|
|-|-|-|
True|	True|	True|
False|	True|	True|
True|	False|	True|
False|	False|	False|

The `not` Boolean Operator:

It takes one argument and returns the opposite result: False for True and True for False. Here it is in a truth table:

A|	not A|
-|-|
True|	False|
False|	True|

#### 11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

```python
# in case of range(10) by default the range starts with 0 and step size is 1
for i in range(10):
    print(i)
```

```python
# in case of range(0,10) the default step size is 1
for i in range(0,10):
    print(i)
```

```python
for i in  range(0, 10, 1):
    print(i)
```

So, it can be seen that there is no difference between range(10), range(0, 10), and range(0, 10, 1).

#### 12. Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop.

```python
# program to print the numbers 1 to 10 using a for loop
for i in range(1,11):
    print(i)
```

```python
# program to print the numbers 1 to 10 using a while loop
i=1
while i<11:
    print(i)
    i+=1
```
